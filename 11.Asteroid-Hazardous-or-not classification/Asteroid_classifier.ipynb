### Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import seaborn as sns
```

### Read the data

```python
df=pd.read_csv('nasa.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

### Drop the Unnecessary Columns

```python
df = df.drop(['Neo Reference ID', 'Name', 'Orbit ID', 'Close Approach Date',  'Epoch Date Close Approach', 'Orbit Determination Date'] , axis = 1)
df.head()
```

### Encode HazardousColumn

```python
hazardous_labels=pd.get_dummies(df['Hazardous'])
hazardous_labels
```

```python
df=pd.concat([df, hazardous_labels], axis=1)
df.head()
```

```python
df=df.drop(['Hazardous'],axis=1)
df.head()
```

```python
df.info()
```

```python
# Observe Unique values in Orbiting Body and Equinox
```

```python
df['Orbiting Body'].value_counts()
```

```python
df['Equinox'].value_counts()
```

```python
#only,there is a single unique value, so these both can be dropped
```

```python
df=df.drop(['Orbiting Body','Equinox'], axis=1)
```

#### Correlation Analysis

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True)
```

```python
#drop these corr-1 columns,take one and drop other as they are identical
```

```python
df = df.drop(['Est Dia in KM(max)', 'Est Dia in M(min)', 'Est Dia in M(max)', 'Est Dia in Miles(min)'
             ,'Est Dia in Miles(max)', 'Est Dia in Feet(min)', 'Est Dia in Feet(max)', 
             'Relative Velocity km per hr', 'Miles per hour', 'Miss Dist.(lunar)', 
             'Miss Dist.(kilometers)', 'Miss Dist.(miles)'], axis = 1)
df.head()
```

```python
# final heatmap
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True)
```

#### Drop the false columns
we need only either true or false for the classification

```python
df.drop([False], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

### Model Building

```python
x=df.drop([True], axis=1)
y=df[True].astype(int)
```

#### Train-Test-Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, random_state=0, test_size=0.3)
```

### XG Boost Classifier
The XGBoost library provides a built-in function to plot features ordered by their importance.

```python
from xgboost import XGBClassifier
from matplotlib import pyplot
from xgboost import plot_importance

xbg_model = XGBClassifier()
xbg_model.fit(x_train, y_train)
plot_importance(xbg_model)
pyplot.show()
```

### Accuracy_score

```python
from sklearn.metrics import accuracy_score,classification_report

predictions = xbg_model.predict(x_test)
acc = accuracy_score(y_test, predictions)
acc
```

```python
# In %
print(str(np.round(acc*100, 2))+'%')
```

```python
print(classification_report(y_test, predictions))
```
